```python
def random_walk(sigma, n_steps):
    x, y = np.zeros(n_steps), np.zeros(n_steps)
    for i in range(n_steps):
        # random angle
        theta = 2 * math.pi * random.random()
        step_size = np.random.normal(0, sigma) #**2
        x[i] = x[i-1] + step_size * np.cos(theta) 
        y[i] = y[i-1] + step_size * np.sin(theta)
        
    plt.figure(figsize=(7, 5))
    plt.title("Random Walk \n(sigma = " + str(sigma) + " | n = " + str(n_steps) + " steps)", fontsize=18)
    
    plt.axvline(0, color='black', linewidth=.5)
    plt.axhline(0, color='black', linewidth=.5)
    plt.plot(x, y)
```

```python
random_walk(.5, 1000)
random_walk(1, 1000)
random_walk(2, 1000)
```

```python
random_walk(.5, 20)
random_walk(1, 400)
random_walk(2, 1600)
```
